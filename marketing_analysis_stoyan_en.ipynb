#Data Cleaning

```python
df.rename(columns={" Income " : "Income"}, inplace=True)    
```

```python
df.Income.sort_values(na_position='first').head(30)        
```

```python
df.iloc[[134]]                                       
```

```python

to_drop=["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "Response", "Complain"]
```

```python

df.drop(to_drop, inplace=True, axis=1)
```

```python
df.iloc[[134]]                             
```

```python
df_clean = df.dropna()                                            
```

```python
df_clean.info()
```

```python
# transform Income column to a numerical
df_clean['Income'] = df_clean['Income'].str.replace('$', '')
df_clean['Income'] = df_clean['Income'].str.replace(',', '').astype('float')
```

```python
df_clean.info()
```

```python
df_clean
```

```python
df_clean['Dt_Customer'] = pd.to_datetime(df_clean['Dt_Customer'])
```
